```python
import nltk
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize , sent_tokenize
```

```python
input = "Hello I am Dheeraj.How are you?"
output1 = word_tokenize(input)
print(output1)
output2 = sent_tokenize(input)
print(output2)
```

```python
input = "Hi"
output1 = word_tokenize(input)
print(output1)
output2 = sent_tokenize(input)
print(output2)
```
